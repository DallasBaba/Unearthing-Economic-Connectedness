# **List of Power plants with higher Emission Output than input.**
#### When a NOx Emission output rate is greater than the input rate, It means the power plant in the process emits more pollutants than the energy it produces.

###### This query returns the Top 5 power plants that emit more NOx pollution than their average electricity input in 2022  in Tennessee, Arkansas, and Mississippi.

```sql
SELECT DISTINCT FacilityCode, PlantName, EmissionOutput,EmissionIutput, UtilityName,StateABB
FROM Hackathon_LH.dbo_UnadjustedEmissions
WHERE EmissionOutput > (SELECT AVG(EmissionIutput) FROM Hackathon_LH.dbo_UnadjustedEmissions)
AND EmissionOutput is not NUll 
AND EmissionIutput is not NUll 
AND StateABB IN ('TN', 'AR', 'MS')
ORDER BY EmissionOutput DESC
LIMIT 5
```

# **List of Plants with lower Emission Output than Emission Output**
#### This means that the power plant's are relatively efficient in converting fuel into electricity while minimizing emissions.**

```sql
SELECT DISTINCT FacilityCode, PlantName, EmissionOutput,EmissionIutput, UtilityName,StateABB
FROM Hackathon_LH.dbo_UnadjustedEmissions
WHERE EmissionOutput < (SELECT AVG(EmissionIutput) FROM Hackathon_LH.dbo_UnadjustedEmissions)
AND EmissionOutput is not NUll 
AND EmissionIutput is not NUll 
ORDER BY EmissionOutput DESC
LIMIT 5
 
```

```sql
SELECT PlantName, UnadjustedSO2, UnadjustedCO2, UnadjustedHg, UnadjustedHeat,
MIN(UnadjustedNOxEmissions) AS UnadjustedNOxEmissions_Min,
AVG(UnadjustedNOxEmissions) AS UnadjustedNOxEmissions_Avg,
MAX(UnadjustedNOxEmissions) AS UnadjustedNOxEmissions_Max
FROM Hackathon_LH.dbo_UnadjustedEmissions
WHERE EmissionOutput > (SELECT AVG(EmissionIutput) FROM Hackathon_LH.dbo_UnadjustedEmissions)
AND EmissionOutput is not NUll 
AND EmissionIutput is not NUll 
GROUP BY PlantName, UnadjustedSO2, UnadjustedCO2, UnadjustedHg, UnadjustedHeat,
ORDER BY UnadjustedNOxEmissions_Max DESC
```
